```python
#importação de bibli
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import os


#func para n fazer dnv
def plota_pivot_table(df, value, index, func, ylabel, xlabel, opcao='nada'):
    if opcao == 'nada':
        pd.pivot_table(df, values=value, index=index,
                       aggfunc=func).plot(figsize=[15, 5])
    elif opcao == 'sort':
        pd.pivot_table(df, values=value, index=index,
                       aggfunc=func).sort_values(value).plot(figsize=[15, 5])
    elif opcao == 'unstack':
        pd.pivot_table(df, values=value, index=index,
                       aggfunc=func).unstack().plot(figsize=[15, 5])
    plt.ylabel(ylabel)
    plt.xlabel(xlabel)
    return None


#pegando todos os arq csv e colocando em uma lista
lista_dataFrames = []
lista_nome = []
diretorio = "F:\\Nova pasta\\CIENCIA DE DADOS\\Support_Exercise_M14\\input"
for arq in os.listdir(diretorio):
    if arq.endswith(".csv"):
        nome_arquivo = "\\" + arq
        nome_df = pd.read_csv(diretorio+nome_arquivo)
        #print("Nome do arq: ", arq)
        lista_dataFrames.append(nome_df)
        x = arq.rfind(".")
        lista_nome.append(arq[:x])
        #print(arq[:x])


#com a lista feita só criar mais 1 loop para fazer com todos os arq csv do mes de JAN  até DEZ e salvando as img no diretório
for z in lista_dataFrames[:3]:
    max_data = z.DTNASC.max()[:7]
    os.makedirs('./figuras/'+max_data)
    plota_pivot_table(z, 'IDADEMAE', 'DTNASC', 'count', 'quantidade de nascimento','data de nascimento')
    plt.savefig('./figuras/'+max_data+'/quantidade de nascimento.png')  
    
    plota_pivot_table(z, 'IDADEMAE', ['DTNASC', 'SEXO'], 'mean', 'media idade mae','data de nascimento','unstack')
    plt.savefig('./figuras/'+max_data+'/media idade mae por sexo.png')

    plota_pivot_table(z, 'PESO', ['DTNASC', 'SEXO'], 'mean', 'media peso bebe','data de nascimento','unstack')
    plt.savefig('./figuras/'+max_data+'/media peso bebe por sexo.png')

    plota_pivot_table(z, 'PESO', 'ESCMAE', 'median', 'apgar1 medio','gestacao','sort')
    plt.savefig('./figuras/'+max_data+'/media apgar1 por escolaridade mae.png')

    plota_pivot_table(z, 'APGAR1', 'GESTACAO', 'mean', 'apgar1 medio','gestacao','sort')
    plt.savefig('./figuras/'+max_data+'/media apgar1 por gestacao.png')

    plota_pivot_table(z, 'APGAR5', 'GESTACAO', 'mean', 'apgar5 medio','gestacao','sort')
    plt.savefig('./figuras/'+max_data+'/media apgar5 por gestacao.png')
```
